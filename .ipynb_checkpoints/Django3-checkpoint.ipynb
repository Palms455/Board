```python
views.py
lass AddItem(View):

	def get(self, request):
		form = ItemForm() создаем экземпляр форм
		image = PhotoForm()
		return render(request, 'items/add_item.html', context = {'form': form, 'image': image }) - передаем в шаблон поля форм

	def post(self, request):
		its = ItemForm(request.POST) - получаем данные в поля форм
		image = PhotoForm(request.POST, request.FILES) 
		if its.is_valid() and image.is_valid(): проверка на валидность данных
			new_item = its.save(commit=False) commit=False - сохранение queryset
			new_item.date_pub=datetime.datetime.now()
			new_item.save() - сохранение в бд
			for f in request.FILES.getlist('photo'): getlist - если в модели определили несколько сохр фото
				data = f.read()
				photo = ProductPhoto(product = new_item)
				photo.photo.save(f.name, ContentFile(data))
				photo.save()

			return redirect(new_item) - переход на страницу созданного объявления
		print(image.errors)
		return render(request, 'items/add_item.html', context = {'form': its, 'image': image})
```
